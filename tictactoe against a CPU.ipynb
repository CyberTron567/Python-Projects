```python
import random

def print_board(board):
    print("\n")
    print(" " + board[0] + " | " + board[1] + " | " + board[2])
    print("---+---+---")
    print(" " + board[3] + " | " + board[4] + " | " + board[5])
    print("---+---+---")
    print(" " + board[6] + " | " + board[7] + " | " + board[8])
    print("\n")

def player_move(board, player):
    while True:
        move = input("Enter your move (0-8): ")
        if move.isdigit() and int(move) < 9 and board[int(move)] == " ":
            board[int(move)] = player
            break
        else:
            print("Invalid move. Please try again.")

def cpu_move(board, player, cpu):
    for i in range(9):
        if board[i] == " ":
            board[i] = cpu
            if check_winner(board, cpu):
                return
            else:
                board[i] = " "
    for i in range(9):
        if board[i] == " ":
            board[i] = player
            if check_winner(board, player):
                board[i] = cpu
                return
            else:
                board[i] = " "
    while True:
        move = random.randint(0, 8)
        if board[move] == " ":
            board[move] = cpu
            break

def check_winner(board, player):
    for i in range(0, 9, 3):
        if board[i] == board[i+1] == board[i+2] == player:
            return True
    for i in range(3):
        if board[i] == board[i+3] == board[i+6] == player:
            return True
    if board[0] == board[4] == board[8] == player:
        return True
    if board[2] == board[4] == board[6] == player:
        return True
    return False

def tictactoe():
    board = [" "] * 9
    players = ["X", "O"]
    random.shuffle(players)
    print("Player 1: " + players[0])
    print("CPU: " + players[1])
    turn = 0
    while True:
        print_board(board)
        if turn % 2 == 0:
            player_move(board, players[0])
            if check_winner(board, players[0]):
                print_board(board)
                print("Player 1 wins!")
                break
        else:
            cpu_move(board, players[0], players[1])
            if check_winner(board, players[1]):
                print_board(board)
                print("CPU wins!")
                break
        if turn == 8:
            print_board(board)
            print("It's a tie!")
            break
        turn += 1

tictactoe()
```
